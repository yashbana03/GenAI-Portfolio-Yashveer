```python
def longestCommonPrefix(strs):
    if not strs:
        return ""
    
    result = ""
    for i in range(len(strs[0])):
        char = strs[0][i]
        
        # Check ALL other strings (strs[1:])
        for word in strs[1:]:
            if i >= len(word) or word[i] != char:
                return result
        
        result += char
    
    return result
list1 = ["flower","flow","flight"] 
print(longestCommonPrefix(list1))
```
