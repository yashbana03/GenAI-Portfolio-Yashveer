```python
import time 
import numpy as np 
```

```python
# creating array from list 
arr_1d = np.array([1,3,4,5,6])
print("1D array :",arr_1d)
```

```python
arr_2d = np.array( [[1,3,4],[5,6,2]])
print("2D array :", arr_2d)
```

```python
##list vs numpy array 

py_list = [1,2,3,4]
print("list multiplication:", py_list*2)

arr = np.array([1,3,34,4])
print("array multi. :" , arr*2)
```

```python
start = time.time() 

py_list = [ i*2 for i in range(1000000)]
print("\n list operation time : " , time.time()-start)


start = time.time()
np_array = np.arange(10000)*2
print("\n list operation time : " , time.time()-start)
```

```python
# creating array from scratch 
zeros = np.zeros([3,4])
print( " zeros array :\n" , zeros)

ones = np.ones([3,4])
print( " Ones array :\n" , ones)

full = np.full((3,3), 6)
print( " full array :\n" , full)

random = np.random.random((2,3))
print( " random array :\n" , random)


sequence = np.arange(0, 10 , 2)
print( "sequesce array :\n" , sequence)
```

```python
# vector , matrix , tensor

vector = np.array ([ 2,4,4])
print(vector,"\n_____" )


matrix = np.array([[1,2,3],[4,5,6]])
print(matrix ,"\n_____")


tensor = np.array([[[1,2] , [
    2,3]],
                   [[4,5] , [6,7]]])
print(tensor,"\n_____")
```
