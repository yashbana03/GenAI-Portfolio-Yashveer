```python
def apply_discount(price, discount):
    # 1. Type checks (int or float)
    if not isinstance(price, (int, float)):
        return "The price should be a number"
    if not isinstance(discount, (int, float)):
        return "The discount should be a number"
    
    # 2. Value checks
    if price <= 0:
        return "The price should be greater than 0"
    if discount < 0 or discount > 100:
        return "The discount should be between 0 and 100"
    
    # 3. Final calculation
    discount_amount = (discount / 100) * price
    final_price = price - discount_amount
    
    return final_price
apply_discount(100, 20) # 80.0  
apply_discount(-100, 20) # "The price should be greater than 0"
apply_discount(100, 150) # "The discount should be between 0 and 100"
apply_discount("100", 20) # "The price should be a number"
apply_discount(74.5, 20.0) # "The discount should be a number"
```
