```python
words = ['tree', 'sky', 'mountain', 'river', 'cloud', 'sun']
def is_word_long(word):
     return len(word)> 4
long= list(filter(is_word_long, words))
long
```

```python
words = ['tree', 'sky', 'mountain', 'river', 'cloud', 'sun']
def is_word_long(word):
     return len(word)> 4
long= list(map(is_word_long, words))
long
```

```python
celsius = [0, 10, 20, 30, 40]

def to_fahrenheit(temp):
    return (temp * 9/5) + 32 > 80

fahrenheit = list(filter(to_fahrenheit, celsius))
print(fahrenheit)
```

```python
celsius = [0, 10, 20, 30, 40]

def to_fahrenheit(temp):
    return (temp * 9/5) + 32

fahrenheit = list(map(to_fahrenheit, celsius))
print(fahrenheit)
```
