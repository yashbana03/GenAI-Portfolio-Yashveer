```python
desserts = ['Cake', 'Cookies', 'Ice Cream', 'Pie', 'Brownies']

# 1. Basic Slicing [start : stop]
# Index 1 se lekar 4 se pehle tak (1, 2, 3)
print(desserts[1:4])  # ['Cookies', 'Ice Cream', 'Pie']

# 2. Slicing with Step [start : stop : step]
numbers = [1, 2, 3, 4, 5, 6]

# Index 1 se shuru, poore list tak, har 2nd element
print(numbers[1::2])  # [2, 4, 6]

# 3. Kuch aur useful slices:
print(numbers[:3])    # [1, 2, 3] (Shuruat se 3 tak)
print(numbers[3:])    # [4, 5, 6] (3 se ant tak)
print(numbers[::-1])  # [6, 5, 4, 3, 2, 1] (List ko reverse karna)
```

```python
# 1. APPEND() - End mein item add karna
numbers = [1, 2, 3, 4, 5]
numbers.append(6)                    # [1, 2, 3, 4, 5, 6]
numbers.append([7, 8])               # [1, 2, 3, 4, 5, 6, [7, 8]] (nested list)

# 2. EXTEND() - Multiple items individually add karna
numbers = [1, 2, 3, 4, 5]
numbers.extend([6, 8, 10])           # [1, 2, 3, 4, 5, 6, 8, 10]

# 3. INSERT() - Specific position pe add karna
numbers = [1, 2, 3, 4, 5]
numbers.insert(2, 2.5)               # [1, 2, 2.5, 3, 4, 5]

# 4. REMOVE() - First occurrence hatana (value se)
numbers = [10, 20, 30, 40, 50, 50]
numbers.remove(50)                   # [10, 20, 30, 40, 50]

# 5. POP() - Index se remove + return karta hai
numbers = [1, 2, 3, 4, 5]
print(numbers.pop(1))                # 2 (return karta hai)
print(numbers.pop())                 # 5 (last element)

# 6. CLEAR() - List ko completely empty karna
numbers = [1, 2, 3, 4, 5]
numbers.clear()                      # []

# 7. SORT() - Original list sort karta hai (in-place)
numbers = [19, 2, 35, 1, 67, 41]
numbers.sort()                       # [1, 2, 19, 35, 41, 67]

# 8. SORTED() - Nayi sorted list banata hai (original unchanged)
numbers = [19, 2, 35, 1, 67, 41]
sorted_list = sorted(numbers)        # Original unchanged

# 9. REVERSE() - List ko ulta karna (in-place)
numbers = [6, 5, 4, 3, 2, 1]
numbers.reverse()                    # [1, 2, 3, 4, 5, 6]

# 10. INDEX() - Item ka first index dhundna
langs = ['Rust', 'Java', 'Python', 'C++']
print(langs.index('Java'))           # 1
```

```python
# 1. Tuple Creation
developer = ('Alice', 34, 'Rust Developer')
print(developer)  # ('Alice', 34, 'Rust Developer')

# 2. Indexing (List jaisa)
print(developer[1])    # 34
print(developer[-2])   # 34 (second last)

# 3. tuple() constructor
name = 'Jessica'
print(tuple(name))     # ('J', 'e', 's', 's', 'i', 'c', 'a')

# 4. 'in' check
langs = ('Python', 'Java', 'C++', 'Rust')
print('Rust' in langs)        # True
print('JS' in langs)          # False

# 5. Unpacking ⭐
name, age, job = developer
print(name, age, job)         # Alice 34 Rust Developer

# 6. Extended Unpacking (Asterisk *)
name, *rest = developer
print(name)  # Alice
print(rest)  # [34, 'Rust Developer']

# 7. Slicing (List jaisa)
desserts = ('cake', 'pie', 'cookies', 'ice cream')
print(desserts[1:3])  # ('pie', 'cookies')

# ❌ IMMUTABLE ERRORS
# developer[0] = 'Bob'     # TypeError!
# del developer[1]         # TypeError!
```

```python
# 1. COUNT() - Kitni baar aaya item?
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
print(programming_languages.count('Rust'))      # 2
print(programming_languages.count('JS'))        # 0

# 2. INDEX() - Pehli baar kahan mila?
print(programming_languages.index('Java'))      # 1
print(programming_languages.index('Python', 3)) # 5 (index 3 se search shuru)

# 3. INDEX() with start+stop
print(programming_languages.index('Python', 2, 5))  # 2

# 4. SORTED() - Nayi sorted list banata hai
numbers = (13, 2, 78, 3, 45, 67, 18, 7)
print(sorted(numbers))  # [2, 3, 7, 13, 18, 45, 67, 78]

# 5. SORTED() with KEY=len (string length se sort)
print(sorted(programming_languages, key=len))
# ['C++', 'Rust', 'Rust', 'Java', 'Python', 'Python']

# 6. SORTED() Reverse
print(sorted(programming_languages, reverse=True))
# ['Rust', 'Rust', 'Python', 'Python', 'Java', 'C++']
```
