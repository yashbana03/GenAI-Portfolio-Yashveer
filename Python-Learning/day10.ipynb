```python
pizza = {
    'name': 'Margherita Pizza',
    'price': 8.9,
    'calories_per_slice': 250,
    'toppings': ['mozzarella', 'basil']
}

print(pizza['name'])                  # 'Margherita Pizza'
pizza['name'] = 'Margherita'
print(pizza['name'])                  # 'Margherita'

print(pizza.get('toppings', []))      # ['mozzarella', 'basil']
print(pizza.keys())                   # dict_keys(['name', 'price', 'calories_per_slice', 'toppings'])
print(pizza.values())                 # dict_values(['Margherita', 8.9, 250, ['mozzarella', 'basil']])
print(pizza.items())                  # dict_items([...])

removed_price = pizza.pop('price', 10)
print(removed_price)                  # 8.9

last_item = pizza.popitem()
print(last_item)

pizza.update({'price': 15, 'total_time': 25})
print(pizza)

pizza.clear()
print(pizza)                          # {}
```

```python
products = {
    'Laptop': 990,
    'Smartphone': 600,
    'Tablet': 250,
    'Headphones': 70,
}

# Print all values
for price in products.values():
    print(price)

# Print all keys
for product in products.keys():
    print(product)

# Or directly over the dictionary
for product in products:
    print(product)

# Print key-value pairs as tuples
for item in products.items():
    print(item)

# Print key and value separately
for product, price in products.items():
    print(product, price)

# Apply 20% discount to all products
for product, price in products.items():
    products[product] = round(price * 0.8)

print(products)

# Loop with enumerate over keys
for index, product in enumerate(products):
    print(index, product)

# Loop with enumerate over values
for index, price in enumerate(products.values()):
    print(index, price)

# Loop with enumerate over items
for index, product in enumerate(products.items()):
    print(index, product)

# Enumerate starting from 1
for index, product in enumerate(products.items(), 1):
    print(index, product)
```
