```python

# code description: This code defines a class called MusicalInstrument with attributes name and instrument_type. It includes a method called play that prints a message indicating that the instrument is fun to play, and another method called get_fact that returns a string with information about the instrument. Two instances of the MusicalInstrument class are created, and the play method and get_fact method are called for one of the instances.
class MusicalInstrument:
    def __init__(self, name, instrument_type):
        self.name = name
        self.instrument_type = instrument_type

    def play(self):
        print(f'The {self.name} is fun to play!')

    def get_fact(self):
        return f'The {self.name} is part of the {self.instrument_type} family of instruments.'


instrument_1 = MusicalInstrument('Oboe', 'woodwind')
instrument_2 = MusicalInstrument('Trumpet', 'brass')

instrument_1.play()
print(instrument_1.get_fact())
```
