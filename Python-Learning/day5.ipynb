```python
# --- Python Integers & Floats Reference ---

# 1. Basic Types
my_int = 56
my_float = 5.4

# 2. Arithmetic Operations
print("Addition:", my_int + my_float)      # Float result: 61.4
print("Subtraction:", my_int - 10)         # 46
print("Multiplication:", my_int * 2)       # 112
print("Division:", my_int / 12)            # Float division: 4.666...
print("Modulo (Remainder):", 56 % 12)      # 8
print("Floor Division:", 56 // 12)         # 4
print("Exponentiation:", 2 ** 3)           # 8

# 3. Type Conversion
print("Float to Int:", int(12.92))         # 12
print("Int to Float:", float(56))          # 56.0
print("String to Int:", int('45'))         # 45
print("String to Float:", float('7.8'))    # 7.8

# 4. Built-in Methods
print("Rounding:", round(4.798))           # 5
print("Rounding (2 decimal):", round(4.253, 1)) # 4.3
print("Absolute Value:", abs(-15))         # 15
print("Power function:", pow(2, 3))        # 8 (same as 2 ** 3)
print("Modular Exponentiation:", pow(2, 3, 5)) # (2**3) % 5 = 3

# Note: Floats can have rounding errors (e.g., 0.1 + 0.2 = 0.30000000000000004)
```

```python
# --- Python Augmented Assignments Reference ---

# Concept: variable <operator>= value 
# Shorthand for: variable = variable <operator> value

# 1. Arithmetic Augmented Assignments
x = 10
x += 5   # x = x + 5  -> 15
x -= 3   # x = x - 3  -> 12
x *= 2   # x = x * 2  -> 24
x /= 4   # x = x / 4  -> 6.0
x //= 2  # x = x // 2 -> 3.0 (Floor Division)
x %= 2   # x = x % 2  -> 1.0 (Modulo)
x **= 3  # x = x ** 3 -> 1.0 (Exponentiation)

# 2. String Concatenation & Repetition
greet = 'Hello'
greet += ' World'  # Concatenation -> 'Hello World'
greet *= 2         # Repetition    -> 'Hello WorldHello World'

# 3. Important Notes
# - Python mein ++ aur -- (C-style increment/decrement) nahi hote.
# - ++x ya --x likhne par syntax error nahi aata, lekin wo increment nahi karte; 
#   wo sirf unary operators ki tarah behave karte hain (e.g., ++5 is still 5).
# - Hamesha x += 1 ka use karein increment karne ke liye.
# - Strings par sirf += aur *= (repetition) kaam karte hain. 
#   Baki ( -=, /=, etc.) string ke saath TypeError dete hain.
```
