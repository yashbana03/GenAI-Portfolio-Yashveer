```python
def greet(fx):
    def mfx(*args, **kwargs):
        print("good morning")
        fx(*args, **kwargs)
        print("have a nice day")

    return mfx
    
@greet
def hello():
    print("hello world")
@greet
def add(a,b):
    print(a+b)
    
    
hello()
add(2,3)


    
    
```
