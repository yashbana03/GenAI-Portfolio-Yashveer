```python
# 1. Comparison Operators (Return True/False)
print(f"3 > 4:  {3 > 4}")   # False
print(f"3 < 4:  {3 < 4}")   # True
print(f"4 == 4: {4 == 4}")  # True

# 2. Conditional Statement Structure
# Indentation (4 spaces) defines the code block
age = 20

if age >= 65:
    print('You are a senior citizen')
elif age >= 18:
    # Yeh block tabhi run hoga agar upar wali condition False ho
    print('You are an adult')
else:
    # Agar koi bhi condition True na ho
    print('You are not an adult yet')

# 3. Important Notes:
# - ':' (colon) condition ke baad zaroori hai.
# - 'pass' keyword ka use empty block ke liye kiya jata hai.
# - if aur else ke beech mein koi aur statement nahi aa sakti (SyntaxError).
```

```python


# 2. Custom Function with Parameters and Return
def calculate_sum(a, b):
    # 'a' aur 'b' parameters hain
    result = a + b
    return result # Value wapas bhejne ke liye return zaroori hai

print(calculate_sum(5, 10)) # Output: 15
```

```python
# 1. Built-in functions
# int(input()) ka use user se number lene ke liye hota hai
user_age = int(input("Enter your age: ")) 
print(f"User is {user_age} years old.")

# 2. Custom Function with Parameters and Return
def calculate_sum(a, b):
    return a + b

# 3. Calling the function
total = calculate_sum(10, 5) # 10 aur 5 arguments hain
print(f"The sum is: {total}")

# Agar return nahi hota, toh value None hoti
def simple_print(msg):
    print(msg)
    
result_none = simple_print("Hello!")
print(f"Result without return is: {result_none}") 


def simple_print(msg):
    print(msg)
    return msg

result_none = simple_print("Hello!")
print(f"Result without return is: {result_none}") 
```
