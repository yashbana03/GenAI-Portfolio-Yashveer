```python
# 1. Comparison Operators (Return True/False)
print(f"3 > 4:  {3 > 4}")   # False
print(f"3 < 4:  {3 < 4}")   # True
print(f"4 == 4: {4 == 4}")  # True

# 2. Conditional Statement Structure
# Indentation (4 spaces) defines the code block
age = 20

if age >= 65:
    print('You are a senior citizen')
elif age >= 18:
    # Yeh block tabhi run hoga agar upar wali condition False ho
    print('You are an adult')
else:
    # Agar koi bhi condition True na ho
    print('You are not an adult yet')

# 3. Important Notes:
# - ':' (colon) condition ke baad zaroori hai.
# - 'pass' keyword ka use empty block ke liye kiya jata hai.
# - if aur else ke beech mein koi aur statement nahi aa sakti (SyntaxError).
```
