```python
import pandas as pd

# 3 rows aur 2 columns waala integer data
data = {
    "Maths": [85, 92, 78],
    "English": [90, 88, 95]
}

df = pd.DataFrame(data)
print("--- Asli DataFrame ---")
print(df)
```

```python
# 1. df.describe() -> Poore data ki summary (count, mean, std, min, max sab ek sath)
print("\n--- 1. df.describe() ---")
print(df.describe())

# 2. df.mean() -> Har column ka Average nikalta hai
print("\n--- 2. df.mean() ---")
print(df.mean())

# 3. df.corr() -> Columns ke beech ka Correlation (rishta) dikhata hai (-1 se 1 ke beech)
print("\n--- 3. df.corr() ---")
print(df.corr())

# 4. df.count() -> Har column mein kitne non-empty (valid) values hain, woh ginta hai
print("\n--- 4. df.count() ---")
print(df.count())

# 5. df.max() -> Har column ki sabse badi value dhoondta hai
print("\n--- 5. df.max() ---")
print(df.max())

# 6. df.min() -> Har column ki sabse choti value dhoondta hai
print("\n--- 6. df.min() ---")
print(df.min())

# 7. df.median() -> Data ko line se lagane par bilkul beech waala number nikalta hai
print("\n--- 7. df.median() ---")
print(df.median())

# 8. df.std() -> Standard Deviation (Data average se kitna bikhra ya door-door hai)
print("\n--- 8. df.std() ---")
print(df.std())
```
